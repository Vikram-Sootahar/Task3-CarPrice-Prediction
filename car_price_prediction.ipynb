```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 300

brands = {
    'Ford':       ['Focus', 'Fiesta', 'Mondeo', 'Kuga', 'Puma'],
    'Volkswagen': ['Golf', 'Polo', 'Passat', 'Tiguan', 'T-Roc'],
    'BMW':        ['3 Series', '5 Series', 'X3', 'X5', '1 Series'],
    'Mercedes':   ['C Class', 'E Class', 'A Class', 'GLC', 'B Class'],
    'Audi':       ['A3', 'A4', 'A6', 'Q3', 'Q5'],
    'Toyota':     ['Corolla', 'Yaris', 'RAV4', 'Camry', 'C-HR'],
    'Vauxhall':   ['Astra', 'Corsa', 'Insignia', 'Mokka', 'Crossland'],
    'Nissan':     ['Qashqai', 'Juke', 'Leaf', 'Micra', 'X-Trail'],
    'Honda':      ['Civic', 'Jazz', 'CR-V', 'HR-V', 'Accord'],
    'Hyundai':    ['i20', 'i30', 'Tucson', 'Kona', 'Santa Fe'],
}

fuel_types    = ['Petrol', 'Diesel', 'Hybrid', 'Electric']
fuel_weights  = [0.45, 0.30, 0.15, 0.10]
transmissions = ['Manual', 'Automatic']
trans_weights = [0.55, 0.45]

rows = []
for _ in range(n):
    brand  = np.random.choice(list(brands.keys()))
    model  = np.random.choice(brands[brand])
    year   = np.random.randint(2010, 2024)
    age    = 2024 - year
    fuel   = np.random.choice(fuel_types, p=fuel_weights)
    trans  = np.random.choice(transmissions, p=trans_weights)
    engine = np.random.choice([1.0,1.2,1.4,1.5,1.6,2.0,2.5,3.0])
    mileage= np.random.randint(1000, 120000)
    owners = np.random.choice([1,2,3], p=[0.60,0.30,0.10])

    base = {
        'Ford':18000,'Volkswagen':22000,'BMW':35000,
        'Mercedes':38000,'Audi':33000,'Toyota':24000,
        'Vauxhall':16000,'Nissan':20000,'Honda':22000,
        'Hyundai':19000
    }[brand]

    price = base
    price -= age * 1200
    price -= mileage * 0.04
    price += engine * 1500
    price += np.random.randint(-2000, 2000)
    if fuel == 'Electric': price += 5000
    if fuel == 'Hybrid':   price += 2000
    if trans == 'Automatic': price += 1500
    if owners == 2: price -= 1500
    if owners == 3: price -= 3000
    price = max(2000, round(price, -2))

    rows.append({
        'Brand':        brand,
        'Model':        model,
        'Year':         year,
        'Fuel_Type':    fuel,
        'Transmission': trans,
        'Engine_Size':  engine,
        'Mileage':      mileage,
        'Owners':       owners,
        'Price_GBP':    price
    })

df = pd.DataFrame(rows)
df.to_csv('uk_car_data.csv', index=False)

print("=" * 55)
print("  UK USED CARS DATASET CREATED")
print("=" * 55)
print(f"  Total Records : {len(df)}")
print(f"  Total Columns : {len(df.columns)}")
print(f"  Brands        : {df['Brand'].nunique()}")
print(f"  Year Range    : {df['Year'].min()} - {df['Year'].max()}")
print(f"  Price Range   : GBP {df['Price_GBP'].min():,} - {df['Price_GBP'].max():,}")
print("\nFirst 5 rows:")
print(df.head().to_string(index=False))
```

```python
print("=" * 55)
print("  STEP 1 — DATASET SHAPE")
print("=" * 55)
print(f"  Rows    : {df.shape[0]}")
print(f"  Columns : {df.shape[1]}")

print("\n" + "=" * 55)
print("  STEP 2 — DATA TYPES")
print("=" * 55)
print(df.dtypes.to_string())

print("\n" + "=" * 55)
print("  STEP 3 — NULL VALUES CHECK")
print("=" * 55)
null_counts = df.isnull().sum()
if null_counts.sum() == 0:
    print("  No missing values found — Data is clean!")
else:
    print(null_counts.to_string())

print("\n" + "=" * 55)
print("  STEP 4 — DUPLICATE ROWS CHECK")
print("=" * 55)
dupes = df.duplicated().sum()
if dupes == 0:
    print("  No duplicate rows found — Data is clean!")
else:
    print(f"  {dupes} duplicate rows found — removing...")
    df = df.drop_duplicates()
    print("  Duplicates removed!")

print("\n" + "=" * 55)
print("  STEP 5 — STATISTICAL SUMMARY")
print("=" * 55)
print(df.describe().round(2).to_string())

print("\n" + "=" * 55)
print("  STEP 6 — BRAND DISTRIBUTION")
print("=" * 55)
print(df['Brand'].value_counts().to_string())

print("\n" + "=" * 55)
print("  STEP 7 — FUEL TYPE DISTRIBUTION")
print("=" * 55)
print(df['Fuel_Type'].value_counts().to_string())

print("\n" + "=" * 55)
print("  STEP 8 — TRANSMISSION DISTRIBUTION")
print("=" * 55)
print(df['Transmission'].value_counts().to_string())

print("\n" + "=" * 55)
print("  STEP 9 — PRICE ANALYSIS BY BRAND")
print("=" * 55)
brand_price = df.groupby('Brand')['Price_GBP'].mean().sort_values(
    ascending=False).round(2)
for brand, price in brand_price.items():
    print(f"  {brand:<15} : GBP {price:,.2f}")

print("\n" + "=" * 55)
print("  STEP 10 — OUTLIER CHECK")
print("=" * 55)
for col in ['Mileage', 'Price_GBP', 'Engine_Size']:
    print(f"  {col:<15} Min: {df[col].min():,.0f}   Max: {df[col].max():,.0f}")

print("\n" + "=" * 55)
print("  DATA CLEANING COMPLETE")
print("=" * 55)
```
